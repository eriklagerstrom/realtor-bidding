```python
df_final = df.drop(df.columns[-3], axis=1)
```

```python
df_final.head()
```

```python
for col in df_final.columns:
    print(col)
    print(df_final[col].isna().sum())
```

```python
df_final_cleaned = df_final.dropna()
```

```python
df_final_cleaned.head()
```

```python
df_final_cleaned['sqm price'] = df_final_cleaned['final price']/df_final_cleaned['square meters']
```

```python
df_final_cleaned.iloc[:, -1] = df_final_cleaned['sqm price'].round(decimals=0)
```

```python
df_final_cleaned['price increase'] = df_final_cleaned['final price'] - df_final_cleaned['asking price']
```

```python
df_final_cleaned['% increase'] = (df_final_cleaned['final price'] / df_final_cleaned['asking price'] -1)*100
```

```python
df_final_cleaned['% increase'] = df_final_cleaned['% increase'].round(decimals=1)
```

```python
df_final_cleaned.describe()
```

```python
df_final_cleaned.to_excel('SalesCleaned12MUppsala.xlsx', index=False)
```

```python
realtors = df_final_cleaned['realtor house'].unique()
```

```python
num_sales = len(df_final_cleaned)
```

```python
total_sales = df_final_cleaned['final price'].sum()
statistic_columns = ['realtor','sold homes', '% of sales', '% of sold homes', 'avg price development', 'average % development']
```

```python
realtor_stats = pd.DataFrame(columns=statistic_columns)
for realtor in realtors:
    tmp_df = df_final_cleaned[df_final_cleaned['realtor house']==realtor]
    sold_homes = len(tmp_df)
    pct_sales = 100*tmp_df['final price'].sum()/total_sales
    pct_sold_homes = 100*sold_homes / num_sales
    avg_price_dev = tmp_df['price increase'].mean()
    avg_price_dev_pct = tmp_df['% increase'].mean()
    df_add = pd.DataFrame(data=[[realtor, sold_homes, pct_sales, pct_sold_homes, avg_price_dev, avg_price_dev_pct]], columns=statistic_columns)
    realtor_stats = realtor_stats.append(df_add, ignore_index = True)
```

```python
realtor_stats = realtor_stats.sort_values(by=['realtor'])
```
